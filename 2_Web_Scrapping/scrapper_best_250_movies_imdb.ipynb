```python
# importing required Libraries
import pandas as pd  # to create dataframe
import requests  # to send the request to the URL
from bs4 import BeautifulSoup  # to get the content in the form of HTML
# import numpy as np  # to count the values (in our case)
import re
```

```python

# assigning the URL with variable name url
url = "https://m.imdb.com/chart/top/?ref_=nv_mv_250"
HEADERS = {
    "User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"
}
response = requests.get(url, headers=HEADERS)

soup = BeautifulSoup(response.content, "html.parser")

ul_element = soup.find(
    "ul",
    class_="ipc-metadata-list ipc-metadata-list--dividers-between sc-a1e81754-0 eBRbsI compact-list-view ipc-metadata-list--base",
)
```

```python
# creating an empty list, so that we can append the values
title_list = []
link_list = []
year_list = []
duration_list = []
rating_list = []
votes_list = []
```

```python
if ul_element:
    # Encontra todos os elementos li dentro do ul
    li_elements = ul_element.find_all("li")
    print(type(li_elements))
    for li in li_elements:
        print(type(li))
        title = li.find("div", class_="ipc-title").text.strip()
        title = re.sub(r'^\d+\.\s*', '', title) if title else ""
        title_list.append(title)

        link = li.find("a", class_="ipc-title-link-wrapper").get("href")
        link = "https://m.imdb.com" + link if link else ""
        link_list.append(link)

        year = li.find_all(
            "span", class_="sc-b189961a-8 kLaxqf cli-title-metadata-item"
        )[0].text.strip()
        year = year if year else ""
        year_list.append(year)

        duration = li.find_all(
            "span", class_="sc-b189961a-8 kLaxqf cli-title-metadata-item"
        )[1].text.strip().replace(" ", "")
        duration = duration if duration else ""
        duration_list.append(duration)

        rating = li.find("span", class_="ipc-rating-star").text.split()[0].strip()
        rating = rating if rating else ''
        rating_list.append(rating)
        
        votes = li.find("span", class_="ipc-rating-star--voteCount").text.split()[0]
        votes = votes.strip().replace("(", "").replace(")", "").replace("&nbsp;", "")# .replace("mi", "00_000")
        # if 'M' in votes:     
        #     votes = int(float(votes.replace('M', ''))*1_000_000)
        # elif 'K' in votes:
        #     votes = int(float(votes.replace('K', ''))*1_000)
        votes = votes if votes else ""
        votes_list.append(votes)

        # print(
        #     f"Filme: {title}\nlink: {link} \nAno: {year}\nDuração: {duration}\nClassificação: {rating}\nVotos: {votes}\n"
        # )


else:
    print("Elemento <ul> não encontrado.")
```

```python
# Criar DataFrame
df = pd.DataFrame(
    {
        "Title": title_list,
        "Year": year_list,
        "Duration": duration_list,
        "Rating": rating_list,
        "Votes": votes_list,
        "Link": link_list,
    }
)
display(df)
```

```python
df.to_csv(path_or_buf='filmes_scraper.csv' ,index=False,)
```
