```python
import re
import pandas as pd  
import requests 
from bs4 import BeautifulSoup  
```

```python
%%time

# assigning the URL with variable name url
url = "https://m.imdb.com/chart/top/?ref_=nv_mv_250"
HEADERS = {
    "User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"
}
response = requests.get(url, headers=HEADERS)

soup = BeautifulSoup(response.content, "html.parser")

ul_element = soup.find(
    "ul",
    class_="ipc-metadata-list ipc-metadata-list--dividers-between sc-a1e81754-0 eBRbsI compact-list-view ipc-metadata-list--base",
)


# creating an empty list, so that we can append the values
title_list = []
link_list = []
year_list = []
duration_list = []
rating_list = []
votes_list = []

if ul_element:
    # Encontra todos os elementos li dentro do ul
    li_elements = ul_element.find_all("li")
    for li in li_elements:

        title = li.find("div", class_="ipc-title").text.strip()
        title = re.sub(r'^\d+\.\s*', '', title) if title else ""
        title_list.append(title)

        link = li.find("a", class_="ipc-title-link-wrapper").get("href")
        link = "https://m.imdb.com" + link if link else ""
        link_list.append(link)

        year = li.find_all(
            "span", class_="sc-b189961a-8 kLaxqf cli-title-metadata-item"
        )[0].text.strip()
        year = year if year else ""
        year_list.append(year)

        duration = li.find_all(
            "span", class_="sc-b189961a-8 kLaxqf cli-title-metadata-item"
        )[1].text.strip().replace(" ", "")
        duration = duration if duration else ""
        duration_list.append(duration)

        rating = li.find("span", class_="ipc-rating-star").text.split()[0].strip()
        rating = rating if rating else ''
        rating_list.append(rating)
        
        votes = li.find("span", class_="ipc-rating-star--voteCount").text.split()[0]
        votes = votes.strip().replace("(", "").replace(")", "").replace("&nbsp;", "")# .replace("mi", "00_000")
       
        votes = votes if votes else ""
        votes_list.append(votes)

        
else:
    print("Elemento <ul> não encontrado.")
```

```python
# Criar DataFrame
df = pd.DataFrame(
    {
        "Title": title_list,
        "Year": year_list,
        "Duration": duration_list,
        "Rating": rating_list,
        "Votes": votes_list,
        "Link": link_list,
    }
)
display(df)
```

Somente os links serão necessários


```python
links = df['Link'].to_list()
links = [url.split('?')[0] for url in links]
```

```python
X = "reviews?spoiler=hide&sort=reviewVolume&dir=desc&ratingFilter="
HEADERS = {
    "User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"
}

data = []

for i in range(len(links)):
    if  (i+1)%25 ==0:
        print(i+1) 
    # print("Filme", i+1)
    
    link = links[i]
    for j in range(10):
        url = link + X + str(j + 1)
        response = requests.get(url, headers=HEADERS)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            # Encontrando a lista de reviews
            review_container = soup.find_all("li", class_="ipl-content-list__item")
            # Lista para armazenar os dados dos review_container
            
            response = requests.get(url, headers=HEADERS)

            for review in review_container:
                
                movie_name = soup.find('h5', class_='title-header').find('a').text.strip()
                movie_name = movie_name.replace("\n","")
                movie_name = movie_name if movie_name else ""
                
                title = review.find("a", class_="title").text.strip()
                title = title if title else ""

                review_text = review.find("div", class_="text").text.strip()
                review_text = review_text if review_text else ""

                rating = (
                    review.find("span", class_="rating-other-user-rating")
                    .text.strip()
                    .split("/")[0]
                )
                rating = rating if rating else ""

                date = review.find("span", class_="review-date").text.strip()
                date = date if date else ""

                data.append(
                    {
                        "movie": movie_name,
                        "title": title,
                        "review": review_text,
                        "rating": rating,
                        "date": date,
                        "link": url,
                    }
                )

df_reviews_scrapper = pd.DataFrame(data)
```
