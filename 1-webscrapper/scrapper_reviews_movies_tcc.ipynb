```python
import pandas as pd  
import requests 
from bs4 import BeautifulSoup  
```

```python
df = pd.read_csv('filmes_scraper.csv')
df
```

```python
links = df['Link'].to_list()
links = [url.split('?')[0] for url in links]
```

```python
X = "reviews?spoiler=hide&sort=reviewVolume&dir=desc&ratingFilter="
HEADERS = {
    "User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"
}

data = []

for i in range(len(links)):
    if  (i+1)%2 ==0:
        print(i+1) 
    # print("Filme", i+1)
    
    link = links[i]
    for j in range(10):
        url = link + X + str(j + 1)
        response = requests.get(url, headers=HEADERS)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            # Encontrando a lista de reviews
            review_container = soup.find_all("li", class_="ipl-content-list__item")
            # Lista para armazenar os dados dos review_container
            
            response = requests.get(url, headers=HEADERS)

            for review in review_container:
                
                movie_name = soup.find('h5', class_='title-header').find('a').text.strip()
                movie_name = movie_name.replace("\n","")
                movie_name = movie_name if movie_name else ""
                
                title = review.find("a", class_="title").text.strip()
                title = title if title else ""

                review_text = review.find("div", class_="text").text.strip()
                review_text = review_text if review_text else ""

                rating = (
                    review.find("span", class_="rating-other-user-rating")
                    .text.strip()
                    .split("/")[0]
                )
                rating = rating if rating else ""

                date = review.find("span", class_="review-date").text.strip()
                date = date if date else ""

                data.append(
                    {
                        "movie": movie_name,
                        "title": title,
                        "review": review_text,
                        "rating": rating,
                        "date": date,
                        "link": url,
                    }
                )

df_reviews_scrapper = pd.DataFrame(data)
```
