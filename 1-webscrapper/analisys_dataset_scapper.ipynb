```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("reviews_movies_scraper.csv")#[['movie', 'title', 'rating']]
df
```

```python
df
```

```python
df['review'] = df['review'].apply(lambda x: x.encode('ascii', 'ignore').decode('ascii'))
print(f"Quantidade de valores vazios em review: {len(df[df['movie'] == ''])}")
print(f"Quantidade de valores vazios em review: {len(df)- len(df.dropna())}")
```

```python
df= df[['movie', 'review',  'rating']]
```

```python
df
```

```python
print("Estatísticas descritivas:")
df.describe()
```

```python
print("\nContagem de valores únicos por coluna:")
for col in df.columns:
    print(f"{col}:")
    print(df[col].value_counts())
    print()
```

```python
df['rating'].info()
```

```python
# Criando um gráfico
plt.figure(figsize=(5, 3))
sns.countplot(x='rating', data=df)
plt.title('Contagem de valores de pontuações')
plt.xlabel('Valores')
plt.ylabel('Contagem')
plt.show()
```
