```python
# Carregar os dados originais, do scrapper
FILE_PATH = "D:/tcc2/guilherme/2_Web_Scrapping/data/reviews_scraper_original.csv"
df = pd.read_csv(FILE_PATH) 
df
```

```python
# Função para detectar a língua de um texto, com tratamento de exceção
def detect_language(text):
    try:
        return detect(text)
    except LangDetectException:
        return None
```

```python
# Aplicar a função de detecção de língua e filtrar os textos em inglês
df['language'] = df['review'].apply(detect_language)
df_english = df[df['language'] == 'en']
df_not_english = df[df['language'] != 'en']
```

```python
print("Linguagens identificadas:")
print(df['language'].value_counts())
print("Comentários em inglês: ",df_english['language'].value_counts().sum())
print("Comentários em outras linguas:",df_not_english['language'].value_counts().sum())
```
